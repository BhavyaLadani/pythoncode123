```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame({"Hours Studied":[1,2,3,4,5,6,7],
                "Exam Score":[50,55,65,70,75,80,85]})
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
x=df.iloc[:,0:1]
y=df.iloc[:,-1]
print(x.shape)
print(y.shape)
```

```python
# x=df[["Hours Studied"]]
# y=df["Exam Score"]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(x,y)
y_pred=lr.predict(x)
print(y_pred)
```

```python
diff=pd.DataFrame({"Actual":y,"Predict":y_pred})
diff
```

```python
print("Coefficeint : ",lr.coef_)
print("Intercept : ",lr.intercept_)
```

```python
from sklearn import metrics
```

```python
print("MAE :",metrics.mean_absolute_error(y,y_pred))
print("MSE : ",metrics.mean_squared_error(y,y_pred))
print("R^2 Score : ",metrics.r2_score(y,y_pred))
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y,color="blue",label="Actual Score")
plt.plot(x,y_pred,color="red",label="Predicte Score")
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.title("Linear Regression")
plt.legend()
plt.grid(True)
plt.show()
```

```python
x=dataset.iloc[:,0:1]
y=dataset.iloc[:,-1]
print(x.shape)
print(y.shape)
```

```python
print(type(x))
print(type(y))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
print(y_pred)
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
print("MAE :",metrics.mean_absolute_error(y_test,y_pred))
print("MSE : ",metrics.mean_squared_error(y_test,y_pred))
print("R2 Score : ",metrics.r2_score(y_test,y_pred))
```

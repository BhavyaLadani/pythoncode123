```python
import pandas as pd
```

```python
df=pd.DataFrame({'Name':["Alice","BOB","Charlie","David"],
                "Score":[88,92,85,90]})
df
```

```python
sorted_df=df.sort_values(by="Score",ascending=False)
sorted_df
```

```python
df.sort_index(ascending=False)
```

```python
df.set_index('Name',inplace=True)
df
```

```python
s=pd.Series([35,67,90,45])
def pass_fail(x):
    return "pass" if x>=50 else "fail"
result=s.apply(pass_fail)
result
```

```python
df=pd.DataFrame({'Python':[60,70,80],
                'Chemistry':[65,75,85],
                'Maths':[70,80,90]})
result=df.apply(sum,axis=1)
result
```

```python
result=df.apply(lambda x:x.mean())
print(result)
```

```python
df['total']=df.apply(lambda x:x.sum(),axis=1)
df
```

```python
def grade(total):
    if total>=250:
        return "A"
    elif total>=200:
        return "B"
    else:
        return "C"
df["Grade"]=df["total"].apply(grade)
df
```

```python
df=pd.DataFrame({"Width":[1,3],"Height":[3,4]})
def get_stats(row):
    return [row["Width"]*row["Height"],2*(row["Width"]+row["Height"])]
df.apply(get_stats,axis=1,result_type="broadcast")
# reduce , expand , broadcast
```

```python
def add_value(x,y,z):
    return x+y+z
df=pd.DataFrame({'A':[1,2],"B":[10,20]})
df1=df.apply(add_value,args=(1,2))
print(df1)
```

```python
def odd_value(x,m):
    return x*m
df=pd.DataFrame({'A':[1,2],"B":[10,20]})
df2=df.apply(odd_value,m=10)
print(df2)
```

```python
# astype()
df=pd.DataFrame({'A':['1','2','3'],'B':['4','5','6']})
print(df.dtypes)
df['A']=df["A"].astype(int)
print(df.dtypes)
```

```python
# replace()
df=pd.DataFrame({'Marks':[80,85,0,90],"Attedance":[75,80,0,85],"Grade":["A","B","F","A"]})
result=df.replace(0,60)
print(result)
result=df.replace([80,75],[90,80])
print(result)
```

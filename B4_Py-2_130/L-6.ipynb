```python
import pandas as pd
```

```python
df1=pd.DataFrame({'A':[1,2],'B':[3,4]})
df2=pd.DataFrame({'A':[5,6],'C':[7,8]})
result=pd.concat([df1,df2],axis=0,join='inner',ignore_index=True)
print(result)
```

```python
df1=pd.DataFrame({'E':[1,2],'B':[3,4]})
df2=pd.DataFrame({'C':[5,6],'D':[7,8]})
result=pd.concat([df1,df2],sort=True)
print(result)
```

```python
df1=pd.DataFrame({'Employee':['A','B','C'],
                 "Departments":['IT','HR','Finance']})
df2=pd.DataFrame({'Employee':['A','B','D'],
                 "Salary":[50000,40000,30000]})
result=pd.merge(df1,df2,on="Employee")
print(result)
# merge(left,right,how="inner",on,left_on,right_on,left_index,right_index)
```

```python
result=pd.merge(df1,df2,on="Employee",how="outer")
print(result)
```

```python
result=pd.merge(df1,df2,on="Employee",how="left")
print(result)
```

```python
df1=pd.DataFrame({'Emp_id':[101,102,103],
                 'Name':["Amit","Riya","John"],
                 "Salary":[50000,60000,55000]})
df2=pd.DataFrame({'Emp_id':[101,102,104],
                 'Name':["Amit","Riya","David"],
                 "Salary":[52000,62000,58000]})
result=pd.merge(df1,df2,on="Emp_id",suffixes=('_left','_right'))
print(result)
```

```python
df1=pd.DataFrame({'Emp_id':[101,102,103],
                 'Name':["Amit","Riya","John"],})
df2=pd.DataFrame({'Id':[101,102,104],
                 "Salary":[52000,62000,58000]})
result=pd.merge(df1,df2,left_on="Emp_id",right_on='Id')
print(result)
```

```python
df1=pd.DataFrame({'Physics':[60,70,80]},
                 index=[101,102,103])
df2=pd.DataFrame({"Name":['Alice','Bob','Charlie']},index=[101,102,103])
result=pd.merge(df1,df2,left_index=True,right_index=True)
print(result)
```

```python
df1=pd.DataFrame({'Physics':[60,70,80],
                 'student_id':[101,102,103]})
df2=pd.DataFrame({"Name":['Alice','Bob','Charlie']},index=[101,102,103])
result=pd.merge(df1,df2,left_on='student_id',right_index=True)
print(result)
```

```python
df=pd.DataFrame({'Product Category':['Electronics','Furnitate','Electronics','Furnitate'],
                'Sales':[200,150,300,100]})
df
```

```python
result=df.groupby('Product Category')
print(result)
```

```python
result=df.groupby('Product Category')['Sales'].sum()
print(result)
```

```python
result=df.groupby('Product Category')['Sales'].mean()
print(result)
```

```python
result=df.groupby('Product Category')['Sales'].median()
print(result)
```

```python
result=df.groupby('Product Category')['Sales'].describe()
print(result)
```

```python
result=df.groupby('Product Category')['Sales'].sample()
print(result)
```

```python
result=df.groupby('Product Category')['Sales'].size()
print(result)
```

```python
result=df.groupby('Product Category')['Sales'].groups
print(result)
```

```python
result=df.groupby('Product Category',axis=0,as_index=True,sort=True)['Sales'].count()
print(result)
```

```python
result=df.groupby('Product Category',axis=0,as_index=False,sort=True)['Sales'].count()
print(result)
```

```python
df=pd.DataFrame({'Region':['East','East','West','West','East','West'],
                'Sales':[100,150,200,120,110,130],
                'Manager':['A','B','C','D','E','F']})
df
```

```python
def sales_range(group):
    return group['Sales'].max()-group['Sales'].min()
sales_data=df.groupby('Region').apply(sales_range)
print(sales_data)
```

```python
df=pd.DataFrame({'Department':['Sales','Sales','HR','HR','IT','IT'],
                "Salary":[50000,60000,40000,45000,70000,75000],
                "Bonus":[5000,6000,2000,2500,8000,8500]})
result=df.groupby('Department').agg('mean')
print(result)
```

```python
df=pd.DataFrame({'Department':['Sales','Sales','HR','HR','IT','IT'],
                "Salary":[50000,60000,40000,45000,70000,75000],
                "Bonus":[5000,6000,2000,2500,8000,8500]})
result=df.groupby('Department').agg(['mean','min','max'])
print(result)
```

```python
df=pd.DataFrame({'Department':['Sales','Sales','HR','HR','IT','IT'],
                "Salary":[50000,60000,40000,45000,70000,75000],
                "Bonus":[5000,6000,2000,2500,8000,8500]})
result=df.groupby('Department').agg({'Salary':['mean','max'],
                                    'Bonus':['sum']})
print(result)
```

```python
df=pd.DataFrame({'Department':['Sales','Sales','HR','HR','IT','IT'],
                "Salary":[50000,60000,40000,45000,70000,75000],
                "Bonus":[5000,6000,2000,2500,8000,8500]})
result=df.groupby('Department').agg(Meansalary=('Salary','mean'),
                                    Maxsalary=('Salary','max'),
                                    TotalBonus=('Bonus','sum'))
print(result)
```

```python
# UNIT :::::  2 - Data Visualization
# Heat Map
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
data=np.random.randint(low=1,high=100,size=(10,10))
print(data)
sns.heatmap(data)
plt.show()
```

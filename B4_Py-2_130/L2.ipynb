```python
data={'calaries':[420,380,390,400],
     'duration':[50,40,45,48]}
df=pd.DataFrame(data)
print(df)
print(df.loc[0])
print(df.loc[1])
print(df['duration'])
print(df['calaries'])
print(df.loc[[0,3]])
```

```python
data={'name':['Alice','Bob','Charie'],
     'Age':[25,30,35],
     'City':['New York','Los Angles','Chicago']}
df1=pd.DataFrame(data)
print(df1)
print(df1.loc[0])
print(df1.iloc[0])
print(df1.loc[0:2])
print(df1.iloc[0:2])
print(df1.loc[(df1.Age>26)&(df1.City=="Chicago")])
```

```python
j={'A':[1,2,3,4],'B':[5,6,7,8]}
dj=pd.DataFrame(j,index=['r1','r2','r3','r4'])
print(dj)
dj.loc['r2':'r4','B']=99
print(dj)
dj['C']=[10,20,30,40]
print(dj)
dj.loc[dj['A']==0,'new_col']='Zero'
print(dj)
```

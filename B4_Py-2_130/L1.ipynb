```python
import pandas as pd
```

```python
a=[1,7,2]
myvar=pd.Series(a,index=['x','y','z'])
print(myvar)
print(type(myvar))
print(myvar[0])
print(myvar[1])
```

```python
a=(1.4,2,3,4,5)
b=pd.Series(a)
print(b)
```

```python
s_name=["ram","shyam","radha","seeta","geeta"]
marks=[50,60,70,80,90]
ans=pd.Series(marks,index=s_name,name='Student Result')
print(ans)
```

```python
color={'day1':400,'day2':380,'day3':430,'day4':370}
ans1=pd.Series(color,name="salary per day")
print(ans1)
print(ans1.size)
print(ans1.dtype)
print(ans1.name)
print(ans1.index)
print(ans1.values)
print(ans1.is_unique)
```

```python
subs=pd.read_csv("subs.csv",squeeze=True)
print(subs)
print(type(subs))
print(subs.describe())
print(subs.min())
print(subs.median())
print(subs.sum())
print(subs[subs>200])
print(subs[subs>200].size)
print(subs[subs==subs.max()].index)
```
